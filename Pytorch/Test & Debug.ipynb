```python


#   tensor batch processing of flatten of CNN

t1=torch.tensor([
    [1,1,1,1],
    [2,2,2,2],                #     <- image 1 of array value
    [1,1,1,1],
    [2,2,2,2]
])

t2=torch.tensor([
    [1,1,1,1],
    [2,2,2,2],                   #     <- image 2 of array value
    [1,1,1,1],
    [2,2,2,2]
])

t3=torch.tensor([
    [1,1,1,1],
    [2,2,2,2],                 #     <- image 2 of array value
    [1,1,1,1],
    [2,2,2,2]
])

t=torch.stack((t1,t2,t3))
print(t,"\n")
print(t.size())
```

```python
# Adding a color channel
t=t.reshape(3,1,4,4)
print(t)
```
